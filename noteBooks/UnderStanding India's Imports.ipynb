```python
# Calculate the median value for each HSCode
median_values = Trade_Imports.groupby(['HSCode'])['value'].median()
median_values
```

```python
# Fill the missing values with median values for corresponding HSCode
Trade_Imports['value'] = Trade_Imports.groupby('HSCode')['value'].apply(lambda x: x.fillna(median_values[x.name]))
```

```python
grouped_data = Trade_Imports.groupby(['year', 'Commodity', 'HSCode'])['value'].sum().reset_index()
pivot_data = grouped_data.pivot_table(index=['year'], columns=['Commodity', 'HSCode'], values='value', aggfunc=np.sum)
```

```python
pivot_data
```

```python
# Sum of all values for each commodity across all years
commodity_sum = pivot_data.sum().reset_index()

# Rename columns for clarity
commodity_sum.columns = ['Commodity', 'HSCode', 'Total_Value']

# Sort by Total_Value in descending order
commodity_sum = commodity_sum.sort_values(by='Total_Value', ascending=False).reset_index()
commodity_sum.drop(['index'],axis=1,inplace=True)
```
