```python
import os
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display


pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
sfe = pd.read_csv("../DataSets/fossil_fuel_operations/country_solid-fuel-transformation_emissions.csv")
ffe = pd.read_csv("../DataSets/fossil_fuel_operations/country_other-fossil-fuel-operations_emissions.csv")
ong_re = pd.read_csv("../DataSets/fossil_fuel_operations/country_oil-and-gas-refining_emissions.csv")
ong_pte = pd.read_csv("../DataSets/fossil_fuel_operations/country_oil-and-gas-production-and-transport_emissions.csv")
cme = pd.read_csv("../DataSets/fossil_fuel_operations/country_coal-mining_emissions.csv")
```

```python
display(sfe.head(1))
display(ffe.head(1))
display(ong_re.head(1))
display(ong_pte.head(1))
display(cme.head(1))
```

```python
cme.info()
```

```python
ong_re = ong_re.drop(['start_time', 'end_time', 'temporal_granularity','created_date', 'modified_date','emissions_quantity_units'],axis=1)
ffe = ffe.drop(['start_time', 'end_time', 'temporal_granularity','created_date', 'modified_date','emissions_quantity_units'],axis=1)
sfe = sfe.drop(['start_time', 'end_time', 'temporal_granularity','created_date', 'modified_date','emissions_quantity_units'],axis=1)
ong_pte = ong_pte.drop(['start_time', 'end_time', 'temporal_granularity','created_date', 'modified_date','emissions_quantity_units'],axis=1)
cme = cme.drop(['start_time', 'end_time', 'temporal_granularity','created_date', 'modified_date','emissions_quantity_units'],axis=1)
```

```python
cme.info()
```

```python
display(sfe.head(1))
display(ffe.head(1))
display(ong_re.head(1))
display(ong_pte.head(1))
display(cme.head(1))
```

```python
median_emissions = sfe.groupby('iso3_country')['emissions_quantity'].median()
sfe['emissions_quantity'].fillna(sfe['iso3_country'].map(median_emissions), inplace=True)

median_emissions = ffe.groupby('iso3_country')['emissions_quantity'].median()
ffe['emissions_quantity'].fillna(ffe['iso3_country'].map(median_emissions), inplace=True)

median_emissions = cme.groupby('iso3_country')['emissions_quantity'].median()
cme['emissions_quantity'].fillna(cme['iso3_country'].map(median_emissions), inplace=True)

median_emissions = ong_re.groupby('iso3_country')['emissions_quantity'].median()
ong_re['emissions_quantity'].fillna(ong_re['iso3_country'].map(median_emissions), inplace=True)

median_emissions = ong_pte.groupby('iso3_country')['emissions_quantity'].median()
ong_pte['emissions_quantity'].fillna(ong_pte['iso3_country'].map(median_emissions), inplace=True)
```

```python
cme_grouped = cme.groupby(['iso3_country','original_inventory_sector','gas']).sum()
display(cme_grouped.head(5))

total_cme = cme.groupby('iso3_country')['emissions_quantity'].sum()
display(total_cme.sort_values(ascending=False))
```

```python
df = pd.concat([ong_re, ong_pte])
```

```python
grouped = df.groupby(['iso3_country'])['emissions_quantity'].sum().reset_index()
display(grouped.sort_values(by='emissions_quantity',ascending=False).reset_index().drop('index',axis=1))
```

```python
ffe_grouped = ffe.groupby(['iso3_country','original_inventory_sector','gas']).sum()
display(ffe_grouped.head(5))

total_ffe = ffe.groupby('iso3_country')['emissions_quantity'].sum()
display(total_ffe.sort_values(ascending=False))
```
