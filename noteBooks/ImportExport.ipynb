```python
# Pivot table for weight/value by year and resource
df_pivot = ResourceTrade.pivot_table(index=['Resource', 'Year'], values='weight/value')

# Pivot table for total weight and value by resource
df_sum = ResourceTrade.pivot_table(index=['Resource'], values=['Weight (1000kg)', 'Value (1000USD)'], aggfunc='sum')

# Calculate weight/value for total weight and value
df_sum['weight/value'] = df_sum['Weight (1000kg)'] / df_sum['Value (1000USD)']
```

```python
df_sum.sort_values(by='weight/value',ascending=False).reset_index()
```
